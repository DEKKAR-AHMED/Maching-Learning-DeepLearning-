```python
# import data 
# to say to pandas that the columns is separated by a tabulation 
# to avoid the double quotes probleme  quoting = 3 
dataset = pd.read_csv("Restaurant_Reviews.tsv",delimiter="\t",quoting = 3)
```

# cleaning The text 

```python
# data preprocessing Cleaning the text instead of having tweet we can have article comentaire books 
#because at the end we will create whats called a bag of words model and this last 
# consist on getting only the relevant word from each review  
```

```python
import  re # to clean the text 
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# corpus cause in NLP can be anything article long text paragraph
corpus = []

for i in range(0,1000):
    ## first step of cleaing text 

    # the cap []mean we don't want to delete the following 
    # the space mean that the deleted char will be replaced by a space 
    review = re.sub("[^a-zA-Z]"," ",dataset["Review"][i]) # to delete number ponctuation ?!...  and keep only letter 

    #### the seconde step of cleaning text 


    review = review.lower() # put the review in lower cases 
    review

    # the third step is to delete the none significat (irrelevent ) word like that,this,...
    #import nltk
    #from nltk.corpus import stopwords
    nltk.download("stopwords") #we download the list of word that if we find them will be deleted 
    review = review.split()
    # to execute our algorithme faster use set 
    #from nltk.stem.porter import PorterStemmer
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set (stopwords.words("english"))]
    review

    ###### four step stemming keep the root of the word we don t need past or future 
    #from nltk.stem.porter import PorterStemmer
    #ps = PorterStemmer()
    # linked all to be a sentence 
    review = " ".join(review)
    corpus.append(review)
```

# creating the Bag of words or sparse matrix 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_features = cv.fit_transform(corpus).toarray()
```
