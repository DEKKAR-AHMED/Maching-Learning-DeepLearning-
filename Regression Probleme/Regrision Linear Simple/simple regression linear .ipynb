```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# SIMPLE LINEAR REGRESSION 

```python
# import the data set 

dataset = pd.read_csv("Salary_Data.csv")
dataset
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
# spliting the data 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# treat the categoriel variable or dummy them 
# scalling the features 
# we dont need to do the feature scalling because we have only 2 feature 
# and we don t need to dummy them cause we dont have categoriel variable 
```

```python
# fitting our training set with Simple Regression final 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# predict the test_training resulat  

```python
predict = regressor.predict(x_test)
```

# visaulisation the training set result 

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,regressor.predict(x_train),color="blue")
plt.xlabel("experience")
plt.ylabel("salaire ")
plt.title("Regression linear")
plt.show()
```

# visualisation the test set result

```python
plt.scatter(x_test,y_test,color="red")
plt.plot(x_train,regressor.predict(x_train),color="blue")
plt.xlabel("experience")
plt.ylabel("salaire ")
plt.title("Regression linear")
plt.show()
```
